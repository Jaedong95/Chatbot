```python
import pandas as pd 
import os 
```

```python
data_path = './Training/label_data'
```

```python
file_list = []

for name in os.listdir(data_path):
    if name.endswith('zip'):
        continue 
    file_list.append(name)
```

```python
file_list[0]
```

```python
file_path = data_path + '/' + file_list[1]
file_path
```

```python
import json

with open(file_path, encoding='utf-8') as f:
    json_data = json.load(f)

json_data.keys()
```

```python
def process_json(json_file):
    with open(json_file, encoding='utf-8') as f:
        try:
            json_data = json.load(f)
        except:
            print(file_path)
            return 
    
    conv_list = [] 
    conv_id = json_data['info'][0]['id']
    context = json_data['info'][0]['annotations']['subject']
    
    for conv in json_data['info'][0]['annotations']['lines']:
        turn_cnt = conv['id']
        intent = conv['speechAct'] 
        text = conv['norm_text'] 
        
        conv_list.append([conv_id, context, turn_cnt, conv['speaker']['id'], conv['speaker']['sex'],\
                          conv['speaker']['age'], intent, text])
    
    return conv_list
```

```python
columns = ['id', 'context', 'turn_cnt', 'speaker_idx', 'speaker_sex', 'speaker_age', 'intent', 'text']
tmp = [None] * len(columns) 

context_df = pd.DataFrame([tmp], columns=columns)
context_df
```

```python
context_df = []

for json_file in file_list:
    data = data_path + '/' + json_file
    try:
        context_df.extend(process_json(data))
    except:
        print(json_file)
        continue 
        
context_df
```

```python
context_df = pd.DataFrame(context_df, columns=columns)
context_df
```
