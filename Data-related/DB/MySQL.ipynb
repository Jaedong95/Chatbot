### Wellness_df 테이블 생성 

```python
sql = 'DROP TABLE wellness_df'

execute_sql(sql)
```

```python
sql = "CREATE TABLE wellness_df(idx int NOT NULL, \
intent varchar(100) NOT NULL, \
keyword varchar(100) NOT NULL, \
utterance varchar(1000) NOT NULL, \
intent_label int NOT NULL, \
intent_keyword varchar(100) NOT NULL,\
ik_label int NOT NULL, \
PRIMARY KEY(idx) \
)"

execute_sql(sql)
```

```python
sql = "ALTER TABLE wellness_df convert to charset utf8;"   # 테이블에 한글 문자 입력

execute_sql(sql)
```

```python
sql = 'DESC wellness_df;'

execute_sql(sql)
```

```python
file = open(os.path.join(data_path, 'Wellness_Conversation_intent_keyword.csv'),'r', encoding='UTF8')
fReader = csv.reader(file)
```

```python
idx_list = []

for idx, line in enumerate(fReader):
    if idx == 0: 
        continue 
    
    try:
        sql = "INSERT INTO wellness_df VALUES('{0}', '{1}', '{2}', '{3}', '{4}', '{5}', '{6}')"\
        .format(line[0], line[1].replace("'", ''), line[2], line[3], line[4], line[5], line[6])
        curs.execute(sql)
        
    except:
        idx_list.append(idx)
        continue     

len(idx_list)
```

```python
sql = "SELECT * FROM wellness_df"

execute_sql(sql)
```
