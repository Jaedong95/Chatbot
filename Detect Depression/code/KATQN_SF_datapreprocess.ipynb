```python
data = pd.read_csv('/content/drive/MyDrive/LAMDA/chatbot/dataset/Wellness_Conversation_keyword2.csv')
data
```

```python
label_score = dict([])
label_score
```

```python
label_score['label_0'] = [70, 10, 20], [70, 20, 10], [80, 10, 10], [90, 5, 5]
label_score['label_1'] = [5, 90, 5], [10, 70, 20], [20, 70, 10], [10, 80, 10]
label_score['label_2'] = [5, 5, 90], [10, 20, 70], [20, 10, 70], [10, 10, 80]
label_score
```

```python
kq_df = data[['utterance', 'keyword_label']]
kq_df
```

```python
kq_df.info()
```

```python
label_score['label_0'][0]
```

```python
random.randint(0, 3)
```

```python
def convert_label(label):
  '''
  label을 입력받아 확률로 반환해주는 함수 
  '''
  random_val = random.randint(0, 3)
  if label == 0: 
    return label_score['label_0'][random_val] 
  elif label == 1: 
    return label_score['label_1'][random_val]
  elif label == 2: 
    return label_score['label_2'][random_val]
```

```python
kq_df2 = kq_df.copy() 
kq_df2.keyword_label = kq_df2.keyword_label.apply(convert_label) 
kq_df2 
```

```python
kq_df.keyword_label = kq_df.keyword_label.apply(convert_label) 
```

```python
kq_df['label_0'] = kq_df['keyword_label']
kq_df['label_1'] = kq_df['keyword_label'] 
kq_df['label_2'] = kq_df['keyword_label'] 
kq_df 
```

```python
del kq_df['keyword_label']
```

```python
kq_df.label_0 = kq_df.label_0.apply(lambda x: x[0]) 
kq_df.label_1 = kq_df.label_1.apply(lambda x: x[1])
kq_df.label_2 = kq_df.label_2.apply(lambda x: x[2]) 
kq_df 
```

```python
kq_df['utterance'] = kq_df['utterance'].apply(lambda x: x.replace('\t', ''))
kq_df
```

```python
kq_count = kq_df.groupby(['label_0', 'label_1', 'label_2']).count()
kq_count.columns = ['count']
kq_count
```
